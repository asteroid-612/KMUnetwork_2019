## 1.4 delay, loss, throughput in network 

성능이 이전보다 좋아졌다. 목차의 제목에 있는 애들을 발전시켰다는 의미이다. 

**패킷을 보낼때 왜 loss와 delay가 일어나는가?**

- queue 에서 대기하는 시간이 좀 있다. (queue delay)
- link 에 실어줘야하는데, 1 bit 씩 싣는다. 이게 시간이 걸린다. transmission delay. (컨베이어 벨트를 생각하면 된다.) 

버퍼가 만약 꽉 차있으면 새로운 패킷을 처리할 수 없다. 그것이 버려지는 것이다. (가장 기본적인 손실이다.)
- queue 에서 쓸데 없는애를 뽑아서 버리면 되지 않느냐? priority queue를 말한다. 

라우터에 패킷 도착해서 다음 라우터로 가는데 걸리는 시간. 

# delay의 구성요소 4가지 
- processing delay : 나가는 링크가 하나밖에 없지만, link들이 여러개 있으면 어디에 넣어야할지를 header를 보고 결정해주는데에 시간이 걸린다. (nodal processing delay = d proc)
- queueing delay : 큐에서 기다리는거 
- **transmission delay : 비트를 링크에 하나씩 싣는데(큐에서 꺼내와서 링크에 올리는 그 순간까지) 걸리는 시간. L/R(패킷 길이를 링크의 대역폭으로 나눠주면 된다.) 링크가 1mbp이다. 1000bit이 있다-> 0.001(1 mili second)**
- propagation delay : 물리적 링크가 전기의 속도로 달려서 다음 라우터로 가는 시간. 

전기는 2x10^8 m/sec 정도의 속도로 간다. 

4개의 딜레이를 전부 더하면 패킷이 라우터에 도착해서 다음 라우터까지 걸리는 시간이 된다. 

일반적으로 국내라면, 200km 짜리 큰 line이 있고, 라우터가 그 끝에 이어져잇다고 하면, 200km/2x10^8 이다. 1/10^3sec, ms-> mili second(10의 -3승)이다. (즉 대한민국에서는 거리가 가까우니까 propagation에 문제는 생기지 않는다.)

전체 속도에 문제를 주는것(일반적인 문제가 되는 친구)은 **transmission delay**이다. 1Mbps = 10^6bps이다. 

### propagation, transmission delay를 계산할 줄 알아야한다. 
- propagation이 문제가 되는 경우 : 거리가 아주 멀어지는 경우에 문제가 된다. 1만 키로미터 정도가 된다면, ... 존나 문제가 될것이다.. 

transmission delay는 대역폭의 문제가 되는것 아닌가? 하드웨어가 존나 빠르게 동작한다면 하여튼 될것인데, 그 신호를 바꿔서 전송하는 것이라서 그 신호 바꾸는 일을 얼마나 빨리 할 수 있느냐가 문제가 된다. 

우리가 손댈 수 있는 부분은 **nodal process delay**이다. **queueing delay**도.

*Caravan analogy(예시!)*
- 동시에 10대의 카라반이 움직여야한다. 
- 톨부스를 통과하는데 1분의 시간이 걸린다고 하면,transmission time이 존나게 길어지면, transmission이 빠르고 propagation이 느린 경우가 더 빨리 모든 차가 도착할것이다. 
- **링크에 올리는데에 시간이 많이 걸리는것이 문제가 된다는 말이다.**

Queueing delay는 사실 일반적으로는 거의 없다. 

**traffic intensity**가 중요하다. 라우터에 몇비트가 1초마다 들어오는지를 나가는 링크의 대역폭으로 나누는것이다. 이게 1에 가까워지면 queue가 쌓여있을 가능성이 매우 높아진다. 이게 낮아지면 큰 문제가 없다. 1을 넘어가면 내가 처리할 수 없는 상황-> packet loss가 일어난다 

average packet arrival time : 어떤 떄에는 안들어오기도 하고 어떤 때에는 몰릴때도 있다는 말임. 그래서 average가 붙는다. 

queueing delay 는 **시간마다 변할 수 있는 유일한 딜레이이다.** -> 수강신청 기간에 괴로워지는 것이다. -> 학교 라우터에 존나많은 패킷이 들어오고, 너무 많으면 손실됨.. 그런 말임. 

큐잉 딜레이를 낮추는 것은 transmission rate를 높여주는것( 돈을 들이세요 )

- 그러면 패킷이 조금씩 들어오고, 그러면 천천히 받아도 손실이 없어진다.

### throughput : 전송률 - end to end 개념이다. 

send와 receieve 사이의 일을 말한다 
- 서버에서 클라이언트까지 1초에 몇비트를 보내고 있는지를 계산하는것이다. 

여기에서 가장 중요한것은 transmission delay이다. 링크에 비트를 올릴때 걸리는 시간을 말한다. 링크들이 여러개 있을 수 있는데 각 링크의 대역폭이 다르다. 그럴때 순간적인 throughput이 있을 수 있고 **average throughput**이 존재할 수 있다. 

instantaneous : 한순간의  | average : 평균적인. 

throughput에 영향을 미치는것은 ? : 더 작은 대역폭을 가지고 있는 링크가 평균이 된다. 

그 작은 대역폭을 가지고 잇는 링크를 **bottleneck link**라고 부른다. 
- 일반적으로 링크가 다양하고, 16개, 20개 링크중 가장 낮은 속도를 가지고 있는 링크가 throughput에 영향을 미친다.

### INTERNET PROTOCOL STACK( 외워야함~ ) - 물리적인 계층
- application layer : 네트워크 application 을 지원
- transport : 작동하는 프로그램에서의 데이터 전송을 할때

프금마들은 link까지만 커버한다. 무선인 physical까지는 우리가 진도를 안나간다.

## 각 레이어는 밑의 레이어의 서비스를 이용해서 윗 레이어를 구성한다. 
- 모든 서비스를 새롭게 구성하기에는 시간이 너무 많이 걸린다. 

*encapsulation* : 윈도우에서 transport까지는 지원한다. application을 만들어서 두 application 사이에 통신이 있었으면 좋겠다고 하면, transport layer로 넘겨주고, 걔가 넘겨주는 작업을 한다. 그 과정에서 필요한 부가적인 정보가 필요해진다. 그것을 위해서 header가 붙는다.(segment)-> 네트워크 계층으로 내려가면 또 헤더 -> 링크 계층으로 내려가면 또 헤더가 붙음 -> **계층을 내려갈때마다 필요한 헤더가 더 붙어서 내려간다.**

화살표 방향으로 메세지가 흐른다. 가장 먼저 물리적인 계층에 들어오고 그다음으로 link, network, transport, application 순으로 올라온다.
