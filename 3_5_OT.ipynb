## 1.2 Network edge
- end sys, access networks, links

네트워크 엣지(호스트나 커뮤니케이션 링크)와 코어(라우터)가 존재한다. 
- 중요성에 의한것이 아니라 구조상의 문제를 다루는 언어적 표현이다. 
- 존나중요한 구글 웹서버도 네트워크 엣지라고 알겟냐 

**네트워크 엣지**
        - 호스트 : 클라이언트와 서버로 이루어진다. 서버들은 데이터 중심이다. 
**access network**
        - 호스트를 처음으로 라우터와 연결시켜주는 애들. 
**network core**
        - 서로 연결된 라우터, 네트워크들의 네트워크를 의미한다. 
        
end system과 edge router는 주거지에서 연결하거나 상업적으로 접근하거나 모바일로 접근하거나 함. 
    - **keep in mind : bandwidth** 이 access network 가 공유된것인지 독점된것인지를 분류해서 생각해야한다.

### Cable network 

미국에서 더 옛날부터 발달했던 방식이다. 미국은 넓어서 전 지역을 공중파로 커버하기가 힘들기 때문에 케이블을 통해서 방송을 많이 했었음. 예전에 미리 땅파서 케이블 만들어놓음. 

케이블 네트워크의 목적이 방송용. 
- 집에서 upstream을 할 일이 없음. 방송을 headend에서 쏴주기만 하면 되는것이다. 
- 인터넷을 사용하려니까 upstream을 해야해서,, : 각자 집에서 다른것을 방문할 수도 있고 그렇게 되어부림 
        - 하나의 선에서 제공하는 대역폭을 나눠 사용해야한다. 
        - **전화선은 전화구까지 고유하게 우리집에만 독점적인(dedicated) 것이다.**
        
채널을 분할해서 사용하는 방법이 필요해졌음. 
- frequency division multiplexing

packet transmission rate(link capacity/ link bandwidth로도 알려져있음)는 패킷을 전송할때 얼마나 시간이 걸리느냐 하는 말이다. 
- bits/(bits/sec) 이므로 sec가 남는다.
