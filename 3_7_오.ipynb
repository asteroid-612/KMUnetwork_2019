## 1.3 Network core

저번시간에 중요하다고 했던 공식말이야,, 좀 있다가 다시 설명해주신다는데? **오쩐다**

네트워크 코어 : 서로 연결된 라우터들의 망이다. 

#### 네트워크 코어의 기능.
**1. 라우팅** : 테이블에다가 어디로 이 패킷이 가야하는지를 작성해줘야한다. 큐를 그 테이블에 따라서 만드므로 그 테이블이 매우 중요하다. 테이블을 만드는 것도 쉬운 일이 아니다. 내용을 만드는것이 라우팅 알고리즘이 하는 일이다. 

**2. 포워딩** : 패킷을 인풋포트로부터 적절한 아웃풋 포트로 전달해주는것이 포워딩이다. **store and forward의 forward와는 다르다.** 빨라야한다.. 링크의 bandwidth가 넓어도 라우터의 속도가 빠르지 않으면 의미가 없다. (라우터가 처리하는 속도가 빨라야한다.)" 


- 패킷을 시작점에서 끝점까지 전달하는 일을 한다. 
- 마지막에는 웹 서버까지 전달시킨다. 전달해주는 방식을 switching 을 하고 그런것이 있다. 
    - packet switching : 큰 파일을 전송하고 싶으면 패킷으로 나눈다. 이후 그것을 각각 나눠서 전송하는 방식을 말한다. 패킷을 전송할때도,, 패킷의 비트 하나씩을 보낸다. (6~70년정도 됐다.)
        - 각 패킷은 full link capacity로 전달이 된다. 
        - 어느 한 순간 그 링크에는 하나의 패킷만이 전달되고 있는 것이다. **(full speed로)** // curcuit에서는 그러지 않는다 
        - 한꺼번에 보낼 수 없을때 packet(os에 단위가 설정되어있음.) 단위로 잘라서 보내는것이다. **store and forward** 먼저 저장을 끝내고 다음 라우터나 호스트에게 보내는것이다. 
        - 라우터가 많으면 많을수록 시간이 많이 걸린다. L/R에다가 라우터의 개수-1개만큼 곱하기 때문이다. 
        - propagation delay 는 없다고 가정한다. 
    > 패킷 스위칭을 사용할때, **손실이 발생할 수밖에 없다.** 라우터는 패킷을 차별하지 않고 오는 순서대로 다음으로 넘기는데, 라우터 사이의 링크가 capacity가 매우 작은 경우, 너무 느리다. 어쩔 수 없이 라우터 안에 있는 queue(buffer)에 패킷들을 다 넣어서 꺼내서 보내야한다.
    > - 들어오는 패킷의 수가 나가는 패킷의 수보다 많을 수 있다. 그 경우, 패킷을 큐에 적재해야하는데, 그러다가 패킷이 누락될 수 있다.(라우터의 메모리는 제한적이기 때문에 그 이상의 패킷이 들어오는 경우 누락이 된다.) drop-packet loss가 생긴다. 
    > - queue에 넣는 딜레이도 존재한다. 큐잉 딜레이 라고 한다. 

라우터가 하는 일은 뭐임? 라우터는 하나의 링크가 아니라 나가는 링크가 여러가지가 있는 경우가 대부분이다. 
- 패킷이 들어왔을때 어디 링크로 보내야 하는지를 어떻게 구분? : 패킷에 적혀있는 도착지점을 보고 어디로 보낼지를 결정해준다. (header value가 보통 가야할 곳이다.)

queue는 보통 전체에 대해서 존재한다. 링크별로 하나씩 있어서 도착점에 맞는 큐에 적재해뒀다가 각 링크가 비면 꺼내서 그 링크로 보낸다. 

테이블이 없으면 보낼 수 있는 방법이 없다. 
- 옥계이.

    - curcuit switching : 전화에서 사용하던 방식을 의미한다. 그래이엄 밸이 전화 만들었을때부터 사용하던 방법. 
        - 여기서는 제한시간 내에 처리가 불가능할것 같으면 아예 받지를 않아서 패킷 스위칭과 달리 패킷의 누락이 일어나지 않는다. 딜레이가 존재하지 않는다. 
        - 가야하는 경로가 정해져있으면, 미리 대역폭을 할당 받아놓는다. 예약받는 메세지를 보내서, 예약이 다 된 경우만 실제로 메세지를 보내는것이다. 그러므로 예약이 불가능한 상태가 오면 메세지를 아예 보낼 수 없게 만들어버림. 
        - 패킷을 단위로 받는것이 아니라 그냥 비트를 받으면 예약한 루트를 통해서 계속 보내는것이다. 링크가 하나밖에 없는것과 같기때문에 패킷 스위칭보다 빠르다. 
        - 예약한 curcuit은 dedicated이므로 남이 사용할 수 없다. 
        - 작은 비트를 보내기 위해서는 조금 맞지 않는다. 
    > FDM vs TDM : 
    > 헤르츠의 크기가 크면 클수록 많은것을 보낼 수 있다. 그 헤르츠를 나눠서 써야하는데, 어떤 것을 기준으로 나눌 것이냐 하는 문제이다. 
    > - FDM(frequency division multiplexing) : 전체에 최대의 frequency가 존재하는데 이것을 4개로 나눈다. 400mb hrtz를 처리하는 링크이다. 그러면 나누면 100mhrtz를 전송할 수 있다. 
    > - TDM(time division multiplexing) : 시간을 나눠서 그 시간에 각 유저가 사용한다. 그러면 250mbps를 사용하게 된다. 시간을 잘 맞춰야하는데(송신기와 수신기가) 현대 송수신기가 이 작업을 해내고 있다.. 
    
    > 두개의 방법을 혼용해서 사용한다. 
        
### 그래서 써킷이 좋냐 패킷이 좋냐? 
> 결론 : 패킷은 " 더 많은 사용자 " 가 사용할 수 있게 만들어준다. 

각각의 유저가 필요한 100kb/s가 필요한데 항상 active하게 사용하는것은 아니다. 전체 시간의 10%만 active하다. 
- curcuit switching을 하게 되면 10명이 링크를 사용할 수 있는데. 
- packet switching을 하게 되면... 11명이 만약 사용한다면? 모두가 동시에 active할 확률이 매우 낮다. 계산을 해보면 35명의 사용자가 있고 10%의 확률로 active해지는데, 0.0004의 확률이 다같이 활동적일 확률이다. 그러므로 35명이 사용할 수 있게 만든다. 

써킷은 예약에도 시간이 걸린다. 패킷은 그런 시간 안필요해서 매우 간단하다. 근데 패킷이 초과적인 혼란이 생겨날 수 있다. (아까 얘기한 누락이 일어날 수있다.) 인터넷에서 패킷을 사용하는데,,, 어떻게 써킷하고 비슷하게 만들 수 있으까 > 음성통화에서 그런게 필요한데.,,
- 이론적으로는 해결 불가능. 
- 링크의 대역폭을 존나 높여서 돈을 들이면 해결이 된다는거에요~

## 인터넷 구조 : 네트워크의 네트워크들 

결과적인 네트워크가 존나 복잡하다. 처음에는 회사에서만 어떻게좀 해보자 이거였는데... 컴퓨터가 존나 많아진거임... 어떻게 연결할테야? 

수백만개의 ISP들이 존재하는데 얘네를 어떻게 다 연결할테냐... 
- 서로를 다 연결한다(미친놈인가?) O(N^2)되는데 규가 봤으면 때렸을듯
- **조그만 isp말고 커다란 global isp가 있어서 다른 애들이랑 연결을 해준다.** 

CUSTOM ISP : 돈을 주고 글로벌 isp에 연결을 하는것이다. (돈을 많이 벌텐데?) 그래서 kt, skt, lg가 생긴거지요. 경쟁자가 생긴다. 

global isp끼리도 선을 연결하지 않으면 내가 lg 유플러스 쓰는 친구에게 전화를 못검... 그래서 나오는것이 
- **peer network**가 생긴다. 서로는 돈 받지 말고 서로에게 오는 packet들을 전송해주기로 하는것이다. customer provider관계가 아니라 동등한 관계로 활동하는것이다. 

약간 계층이 있다. 

**IXP** : 건물 안에서는 라우터들을 연결하는 것이 존재한다. internet exchange point..물리적 장치들이 존재한다는 말이다. 

regional net이 또 존재해서 중간단계의 isp도 존재한다. 

#### 최근에 일어난 일 : 대기업들의 데이터센터끼리 데이터를 옮길 일이 생길 수 있다. 그래서 데이터 센터를 여러 나라에 만들어서 물리적으로 가까운 데이터 센터에 연결을 시켜주는 것이다. 데이터가 하여튼 존나 왔다리 갔다리를 해야한다 이말이야. 상업용 isp 들을 이용하려니까 돈이 존나 들고 느린것이다. (다른 서비스랑도 공유해서 사용해야하니까.) 그래서 구글은 기분이 나빠서 새로운 망을 만들었다. CONTENT PROVIDER NETWORK- > 걍 필요한데에 돈 존나 많이 들여서 케이블 설치해서 구글만의 망을 만든다. 독점적으로 사용하니까 데이터도 많이 처리가 가능함.
